```python
import numpy as np
from numpy import linalg as LA
from matplotlib import pyplot as plt
```

# HW2-Chebyshev Method for Second Order Linear ODE

## Part1: Discretization

The equation to be discretized in this part is:$$ u''(t) + a u'(t) + b u(t) = f(t), $$ we will use chebyshev polynomials to construct vandermonde matrices and sample our value on the chebyshev nodes. Coefficients $a$ and $b$ are considered as arbitrary constants and the initial conditions are given as:$u(0) = 1$ and $u'(0) = 0$

The domain of the problem is from $t=0$ to $t=1$, so we have to apply the affine mapping to compute the chebyshev nodes. The mapping relationship is given as:
$$ t_i = \frac{0+1}{2}+\frac{1-0}{2}x_i,$$ where $$x_i = \cos(\frac{(2i-1)\pi}{2n}i), \quad i = 1 \dotsb n ,$$ is the standard chebyshev nodes.

Chebyshev Polynomial matrix as:$$\begin{bmatrix} 1 & x_0 & T_2(x_0) & T_3(x_0) & \dots & T_n(x_0) \\ 1 & x_1 & T_2(x_1) & T_3(x_1) & \dots & T_n(x_1) \\ 1 & x_2 & T_2(x_2) & T_3(x_2) & \dots & T_n(x_2) \\ \vdots \\ 1 & x_{n-1} & T_2(x_{n-1}) & \dots & T_{n-1}(x_{n-1}) & T_n(x_{n-1})\\ 1 & x_{n} & T_2(x_n) & \dots & T_{n-1}(x_n) & T_n(x_n) \end{bmatrix},$$ where $x \in (-1,1)$. Both chebyshev polynomial and its differentiation are computed by recursion, and when differentiating operator is needed, we apply the relation $$L^{(n)} = T^{(n)}T^{-1},$$ where $n$ is the order of derivative. In the matrix, considering $x \in (-1,1)$, so we have to map the nodes back to the standard interval, the relationship is given as: $$x_i = t_i\frac{2(1+0)}{1-0}+\frac{1+0}{1-0},$$ and the coefficient $\frac{2(1+0)}{1-0}$ jumps out during each differentiation. 

The mapping and the chebyshev matrix construction code as follows:

```python
'''The way of defining chebyshev nodes comes frm Wikipedia and the method of constructing chebyshev vandermonde 
matrices comes from the class.'''
def affine(a,b,n):  #affine mapping
    x = list(np.linspace(1,2*n-1,n))
    x.reverse()
    return (a+b)/2 + (b-a)/2*np.cos(np.asarray(x)*np.pi/2/n)

def cheby_deri_M(x,n = None):  #compute the chebyshev polynomial matrix and its detrivative matrix     
    if n is None:              #if n is specfied, then output is only for the chosen size
        n = len(x)
    #####___chebyshev_M____#######
    T = np.ones((len(x), n))
    if n > 1:
        T[:,1] = x
    for i in range(2,n):
        T[:,i] = 2*x*T[:,i-1] - T[:,i-2]  
    ######___derivatives____####### 
    dT = np.zeros_like(T)   #define zeros matrix of equal size
    dT[:,1] = 1
    dT[:,2] = 4*x              #first two columns computed by differentiating directly
    ddT     = np.zeros_like(T)
    ddT[:,2]= 4
    
    for i in range(3,n):       #n>=2, the rest derivative computed by recursion
        dT[:,i]  = i*(2*T[:,i-1]  + dT[:,i-2]/(i-2))
        ddT[:,i] = i*(2*dT[:,i-1] + ddT[:,i-2]/(i-2))
        
    return [T, dT, ddT]   #p = M[0], p' = M[1], p'' = M[2]
```

We will solve this ODE by chebyshev method and test its accuracy by manufacturing solutions. In the code, we build a class containing the test function along with its derivatives, and the manufacturing function we choose is $f(x) = e^{kx\sin(x)}$, which has the corresponding initial conditions as the problem set. Like in class, we define our initial condition as pairs in the code, the Dirichlet initial condition as $(deri = 0 , u(0) = 1)$, and the Neumann initial conditions as $(deri = 1 , u'(0) = 0)$. Because the whole boundary conditions should enclose the boundary of the problem, so we reconsider the Neumann initial condition as another boundary condition enfored on the right boundary, and in matrix, this is simply by replacing the last row to the corresponding row in the first-derivative matrix. For the forcing function,  we take the linear combination of the manufacturing function and its derivatives, in the case of testing our numerical solver.

Any linear ordinary differential equations can be written as the form as $$L(u(t)) = f(t),$$
where the L is a matrix for computing derivatives. If $$L(u(t)) = a_1A1(u(t)) + a_2A2(u(t)) + \dots, $$ then $$L = a_1A1 + a_2A2 + \dots,$$where $A_i$ could be differential operators calculating any orders of derivative.

```python
'''this solver code is based on the class example, with some modification'''
class exact_expxsin:              # build a class of test functions with corresponding coefficients
    def __init__(self, k=1):
        self.k = k                 # user defined coefficient-1
        
    def u(self, x):                # the original_function
        return np.exp(self.k*x*np.sin(x))
    def du(self, x):
        return self.k * np.exp(self.k*x*np.sin(x)) * (x*np.cos(x)+np.sin(x))
    def ddu(self, x):
        return self.k*(self.k*(np.exp(self.k*x*np.sin(x)))*(np.sin(x)+x*np.cos(x))**2+np.exp(self.k*x*np.sin(x))*(2*np.cos(x)-x*np.sin(x)))
    
    
def solver(n, a , b, rhs, IC1, IC2): # number of points, coefficients, right hand side function, initial conditon-1, IC-2
    
    x = affine(0, 1, n)     # our domain is [0,1], and mapping it to (-1,1) as standard chebyshev points
    T = cheby_deri_M(2*x-1)       # compute chebyshev matrices 
     
    L  = 4*T[2]+2*a*T[1]+b*T[0]  # the differential operator
    
    rhsf = rhs(x)              # the forcing function
    
    for i,deriv,func in [(0, *IC1),(-1,*IC2)]:  #consider the Neumann boundary condition enfored at the right boundary
        L[i] = T[deriv][0]
        rhsf[i] = func(0)
        
    return x, L.dot(LA.inv(T[0])), rhsf

a = 1   # coefficients_a in the ODE  
b = 1   # coefficients_b in the ODE
n = 50  # number of the points 
et = exact_expxsin(1)
t, L, rhs = solver(n, a,b, lambda x: et.ddu(x)+a*et.du(x)+b*et.u(x), IC1=(0,et.u), IC2=(1,et.du))



uu = LA.solve(L, rhs)
tt = np.linspace(0,1,n*4)     #finer points
plt.figure()
plt.plot(t, uu, 'o')
plt.plot(tt, et.u(tt))        #real solution
plt.plot(tt, cheby_deri_M(tt*2-1)[0][:,:len(t)].dot(LA.solve(cheby_deri_M(2*t-1)[0], uu)))  # sample the approximating value on finer grids
plt.grid()
```

From the figure above, with the chebyshev vandermonde matrix computed on the chebyshev nodes, avoiding the ill-conditioning as well as the Runge's effect, we get a relatively good result.

## Part2: Convergence Test

```python
'''this test code is based on the class example, with some modification'''
def error(n, dis, sol):    # n is the resolution, dis is the discretization regime used and sol is the manufacturing solution class
    x, L, f = dis(n, 1, 1, lambda x: sol.ddu(x)+a*sol.du(x)+b*sol.u(x), IC1=(0,sol.u), IC2=(1,sol.du))
    u = LA.solve(L, f)
    return LA.norm(u - sol.u(x), np.inf)    #compute the infinite norm of the error

tt = np.arange(10,100,5)                   #resolution level           
errors = [error(n, solver, et) for n in tt] #for each resolution level, compute the error of infinite norm
plt.figure()
plt.semilogy(tt, errors, 'o', label='numerical')

for p in range(1,5):
    plt.semilogy(tt, 1/tt**(p), label='$n^{-%d}$'%p)
plt.xlabel('n')
plt.ylabel('error')
plt.legend(loc='lower left')
plt.tight_layout()
```

From the figure above, we can approximately see this discretization has the accuracy of order 2, which is not accurate as the example in class. The reason for this may come from the choice of enforcing Neumann initial condition.

## Part3: Dynamics of the Homogeneous Equation 

After ensuring the accuracy in Part1, in this part, we will not bring manufacturing solutions in for the testing. Instead, we will consider the extreme cases of $a$ and $b$ to discover the different dynamics of the solution. 


```python
count = 0
for a in [-1,1,1000]:
    for b in [-1,1,1000]:
        count = count + 1          #subplot indexes
        n = 100   
        t, L, rhs = solver(n, a,b, lambda x:0*x, IC1=(0,lambda x:0*x+1), IC2=(1,lambda x:0*x))
        uuu = LA.solve(L, rhs)     
        plt.subplot(3,3,count)
        TITLE = 'b=' + str(b) + ',' + 'a=' + str(a)      #variable titles
        plt.title(TITLE)
        plt.plot(t, uuu, 'o')
        plt.grid()
        plt.tight_layout()
    
  
```

From the figures above, we can clearly see two regimes of the ODE solution $u(t)$. For $a \approx b$, and $a >> b$, the function $u(t)$ is monotonous whether it is increasing or decreasing depends on the sign of $b$. While for $b >> a$, the solution is oscillating.
